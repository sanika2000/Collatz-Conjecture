# Collatz Conjecture
Collatz Conjecture is a mathematical problem in which an integer if odd is muliplied by 3 plus 1
and if it is even then it is divided by 2 and this sequence converges to 1 at some point of time.

I have calculated collatz conjecture for numbers between 0 and 10000 and plotted the number of iterations for each range(divided into 10 ranges-0 to 1000,1000-2000 and so on upto 9000-10000)

```python
import random
import matplotlib.pyplot as plt;
count = []
numbers = []
range1 = [];
range2 = [];
range3 = [];
range4 = [];
range5 = [];
range6 = [];
range7 = [];
range8 = [];
range9 = [];
range10 = [];
count_iterations1 = [];
count_iterations2 = [];
count_iterations3 = [];
count_iterations4 = [];
count_iterations5 = [];
count_iterations6 = [];
count_iterations7 = [];
count_iterations8 = [];
count_iterations9 = [];
count_iterations10 = [];
for i in range(1,50000):
    count_iterations = 1;
    value = random.randint(1,10000);
    if value<1001:
        range1.append(value);
    else:
        if value in range(1001,2001):
            range2.append(value);
        else:
            if value in range(2001,3001):
                range3.append(value);
            else:
                if value in range(3001,4001):
                    range4.append(value);
                else:
                    if value in range(4001,5001):
                        range5.append(value);
                    else:
                        if value in range(5001,6001):
                            range6.append(value);
                        else:
                            if value in range(6001,7001):
                                range7.append(value);
                            else:
                                if value in range(7001,8001):
                                    range8.append(value);
                                else:
                                    if value in range(8001,9001):
                                        range9.append(value);
                                    else:
                                        range10.append(value);
                
    numbers.append(value);
    value1 = value;
    while value1!=1:
        if value1%2 == 0:
            value1 = value1/2;
        else:
            value1 = (value1 * 3)+1;
        count_iterations = count_iterations + 1;
    if value<1001:
        count_iterations1.append(count_iterations);
    else:
        if value in range(1001,2001):
            count_iterations2.append(count_iterations);
        else:
            if value in range(2001,3001):
                count_iterations3.append(count_iterations);
            else:
                if value in range(3001,4001):
                    count_iterations4.append(count_iterations);
                else:
                    if value in range(4001,5001):
                        count_iterations5.append(count_iterations);
                    else:
                        if value in range(5001,6001):
                            count_iterations6.append(count_iterations);
                        else:
                            if value in range(6001,7001):
                                count_iterations7.append(count_iterations);
                            else:
                                if value in range(7001,8001):
                                    count_iterations8.append(count_iterations);
                                else:
                                    if value in range(8001,9001):
                                        count_iterations9.append(count_iterations);
                                    else:
                                        count_iterations10.append(count_iterations);
    count.append(count_iterations);

plt.scatter(numbers,count,marker=".",c ="black");
plt.xlabel("Number");
plt.ylabel("No of iterations");
```
